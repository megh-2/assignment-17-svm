```python
x= forest_fires.iloc[:,0:28]
y=forest_fires.iloc[:,28]
```

```python
x
```

```python
y
```

```python
# Normalisation function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
# SVM model


svc = SVC(kernel = "linear",random_state=30,gamma=0.1,C=1.0)


svc.fit(x_train,y_train)
```

```python
pred_l= svc.predict(x_test)
```

```python
np.mean(pred_l ==y_test)   ## accuracy = 89%
```

```python
# kernel =  rbf
svc = SVC(kernel = "rbf",random_state=30,gamma=0.1,C=1.0)


svc.fit(x_train,y_train)
```

```python
pred_rbf  = svc.predict(x_test)
np.mean(pred_rbf ==y_test)   ## accuarcy = 71%
```

```python
# kernel  = poly
svc = SVC(kernel = "poly",random_state=30,gamma=0.1,C=1.0)


svc.fit(x_train,y_train)
```

```python
pred_poly = svc.predict(x_test)
np.mean(pred_poly ==y_test)   ## accuarcy = 71%
```

Grid search CV

```python
# rbf
clf = SVC()
param_grid = [{"kernel":["rbf"],'gamma':[0.1],"C":[1.0],"random_state" :[30]}]
```

```python
gsv = GridSearchCV(clf,param_grid,cv= 10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_,gsv.best_score_
```

```python
# testing
clf = SVC(C =1.0,gamma = 0.1)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred) *100
print("Accracy = ",acc)
confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# kernel = poly
svc= SVC()
parma_grid = [{'kernel' : ["poly"],'random_state':[30],'gamma':[0.1],'C':[1.0]}]
gsv = GridSearchCV(svc,parma_grid,cv=10)
gsv.fit(x_train,y_train)
gsv.best_params_ , gsv.best_score_
```

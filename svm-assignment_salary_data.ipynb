```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
enc = LabelEncoder()
for i in string_columns:
    train_data[i] = pd.DataFrame(enc.fit_transform(train_data[i]))
    test_data[i] = pd.DataFrame(enc.fit_transform(test_data[i]))
```

```python
train_data
```

```python
len(train_data.columns)
```

```python
x_train = train_data.iloc[:,0:13]
y_train = train_data.iloc[:,13]
x_test = test_data.iloc[:,0:13]
y_test = test_data.iloc[:,13]
```

```python
# normalisation
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test = norm_func(x_test)
```

# svm  model

```python
x_train,x_test,y_train,y_test = train_test_split(x_train,y_train,test_size = 0.3)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python

clf = SVC(kernel = "linear",random_state=30,gamma=0.1,C=1.0)


clf.fit(x_train,y_train)
```

```python
pred = clf.predict(x_test)
```

```python
np.mean(pred ==y_test)   ## accuracy = 80%
```

```python
# kernel =  rbf
clf = SVC(kernel = "rbf",random_state=30,gamma=0.1,C=1.0)


clf.fit(x_train,y_train)
```

```python
pred_rbf  = clf.predict(x_test)
np.mean(pred_rbf ==y_test)   ## accuarcy = 82%
```

```python
# kernel  = poly
clf = SVC(kernel = "poly",random_state=30,gamma=0.1,C=1.0)


clf.fit(x_train,y_train)
```

```python
pred_poly = clf.predict(x_test)
np.mean(pred_poly == y_test)   # acuracy=81%
```

```python
# testing
clf = SVC(C =1.0,gamma = 0.1)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred) *100
print("Accracy = ",acc)
confusion_matrix(y_test,y_pred)
```
